
# Base de données des villes de France

## Création de la base de données

Créer une base de donnée vide (Bdd_villes)sur "DB Browser (SQLite)"

Créer les tables de travail grâce aux requêtes suivantes

```sql
CREATE TABLE villes_france(
    "ville_id" INTEGER PRIMARY KEY AUTOINCREMENT,
    "ville_departement" text ,
    "ville_slug" text ,
    "ville_nom" text ,
    "ville_nom_simple" text ,
    "ville_nom_reel" text ,
    "ville_nom_soundex" text ,
    "ville_nom_metaphone" text ,
    "ville_code_postal" text ,
    "ville_commune" text ,
    "ville_code_commune" text ,
    "ville_arrondissement" integer,
    "ville_canton" text ,
    "ville_amdi" integer,
    "ville_population_2010" integer,
    "ville_population_1999" integer,
    "ville_population_2012" integer,
    "ville_densite_2010" integer,
    "ville_surface" float ,
    "ville_longitude_deg" float  ,
    "ville_latitude_deg" float  ,
    "ville_longitude_grd" text,
    "ville_latitude_grd" text,
    "ville_longitude_dms" text,
    "ville_latitude_dms" text,
    "ville_zmin" integer,
    "ville_zmax" integer);
```

et

```sql
CREATE TABLE departement(
    departement_id INTEGER PRIMARY KEY AUTOINCREMENT,
    departement_code text ,
    departement_nom text ,
    departement_nom_uppercase text ,
    departement_slug text,
    departement_nom_soundex text);
```    

Récupérer les fichiers "Departement_insert.sql" et "Villes_insert.sql" dans le dossier "Data_Td_villes"du repository.

Afin d'insérer les données faites :
Fichiers -> importer -> Base de données a partir du fichier csv.

Selectionner le fichier puis répondre non a la question :"voulez vous créer etc..."

Faites l'opération pour les deux fichiers.

## Exercices (requêtes SQL)

**Créer une vue pour chacune des requêtes suivantes que vous nommerez "Req1", "Req2" etc...** <br>

Vous pouvez consulter le memo Sql dans le dossier "Informations_diverses"


* 1) Obtenir la liste des 10 villes les plus peuplées en 2012<br>
<br>
* 2) Obtenir la liste des 50 villes ayant la plus faible superficie<br>
<br>
* 3) Obtenir la liste des départements d’outres-mer, c’est-à-dire ceux dont le numéro de département commencent par “97”<br>
<br>
* 4) Obtenir le nom des 10 villes les plus peuplées en 2012, ainsi que le nom du département associé<br>
<br>
* 5) Obtenir la liste du nom de chaque département, associé à son code et du nombre de commune au sein de ces département,      en triant afin d’obtenir en priorité les départements qui possèdent le plus de communes<br>
<br>
* 6) Obtenir la liste des 10 plus grands départements, en terme de superficie<br>
<br>
* 7) Compter le nombre de villes dont le nom commence par “Saint”<br>
<br>
* 8) Obtenir la liste des villes qui ont un nom existants plusieurs fois, et trier afin d’obtenir en premier celles dont le      nom est le plus souvent utilisé par plusieurs communes<br>
<br>
* 9) Obtenir en une seule requête SQL la liste des villes dont la superficie est supérieur à la superficie moyenne<br>
<br>
* 10) Obtenir la liste des départements qui possèdent plus de 2 millions d’habitants<br>
<br>
* 11) Remplacez les tirets par un espace vide, pour toutes les villes commençant par “SAINT-” (dans la colonne qui contient les noms en majuscule)<br>
