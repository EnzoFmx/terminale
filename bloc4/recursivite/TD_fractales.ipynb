## Fractales et récursivité

> D'après wikipedia, une [fractale](https://fr.wikipedia.org/wiki/Fractale) présente des structures similaires à toutes les échelles. Par exemple un [triangle de Sierpiński](https://fr.wikipedia.org/wiki/Triangle_de_Sierpi%C5%84ski) est un triangle composé de trois triangles de Sierpiński, eux-mêmes composés de trois triangles de Sierpiński, eux-mêmes composés de... etc...<br>
![Sierpinski](img/sierpinski.png)
<br> Autrement dit, pour dessiner la structure fractale "triangle de Sierpiński", il faut dessiner à une échelle réduite trois triangles de Sierpiński<br>   
$\Longrightarrow$ On remarque une proximité très forte entre une fractale et la notion de récursivité.
